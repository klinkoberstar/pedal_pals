```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Load the dataset
file_path = "/Users/jenniferyeaton/Documents/Coursework/Autumn_2023/Math_for_ML/project/pedal_pals/rs_million.csv"
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Preprocessing
# Convert 'start_time' to datetime and extract relevant features
data['start_time'] = pd.to_datetime(data['start_time'])
data['hour_of_day'] = data['start_time'].dt.hour
data['day_of_week'] = data['start_time'].dt.hour
data['month_year'] = data['start_time'].dt.to_period('M')
data['year'] = data['start_time'].dt.year
#data['quarter'] = data['start_time'].dt.to_period('Q')
data['day'] = data['start_time'].dt.day_of_week
data['hour'] = data['start_time'].dt.hour
data['month'] = data['start_time'].dt.month
data['user_type'] = data['user_type'].apply(lambda x: 1 if x == 'subscriber' else 0)
data['gender'] = data['gender'].apply(lambda x: 1 if x == 'female' else 0)
data['trip_duration_min'] = data['trip_duration']/60
data = data.dropna(subset=['from_latitude', 'from_longitude'])

# Encoding categorical variables 'user_type'
#label_encoder = LabelEncoder()
#data['user_type_encoded'] = label_encoder.fit_transform(data['user_type'])
#data['gender_numeric'] = label_encoder.fit_transform(data['gender'])

data.head()
```

```python
# Counting the number of NaNs in each column
missing_values = data.isna().sum()

# Displaying the columns with missing values and their counts
missing_values[missing_values > 0]
```

```python
# DONT PLAN TO REPORT THESE FINDINGS: KNN training on full dataset from 2014 - 2019 and test on 2019
# Hypothesis: Will perform well if we forecast 2019 data using 2019 data...

# Selecting relevant columns
X_allyears = data[['from_latitude', 'from_longitude', 'user_type', 'hour_of_day', 'gender', 'day_of_week', 'month', 'year']]
#X_allyears = data[['hour_of_day', 'from_station_id', 'user_type_encoded','gender_numeric', 'birth_year']]
y_allyears = data['trip_duration_min']

# Splitting the dataset into training and testing sets
X_train_allyears, X_test_allyears, y_train_allyears, y_test_allyears = train_test_split(X_allyears, y_allyears, test_size=0.2, random_state=42)

# Training the KNN model
knn = KNeighborsRegressor(n_neighbors=100)
knn.fit(X_train_allyears, y_train_allyears)

# Making predictions
y_pred_allyears = knn.predict(X_test_allyears)

# Predicting and evaluating the model
mse_allyears = mean_squared_error(y_test_allyears, y_pred_allyears)
rmse_allyears = np.sqrt(mse_allyears)

print("Root Mean Squared Error of the Model trained and tested on the full data: ", rmse_allyears)
```
